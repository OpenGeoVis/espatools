```python
# Import the espatools module for raster I/O
import espatools
# Use matplotlib for simple plotting
import matplotlib.pyplot as plt
```

```python
xmlfile = 'LC081210652018073101RT-SC20180802013328/LC08_L1TP_121065_20180731_20180731_01_RT.xml'
# xmlfile = 'LC081210652018073101RT-SC20180802013328/simple.xml'
reader = espatools.RasterSetReader(filename=xmlfile)
```

```python
raster = reader.Read()
bands = raster.bands
```

```python
b = bands.get('sr_aerosol')

plt.figure(figsize=(10,10))
plt.imshow(b.data)
plt.title(b.name)
plt.savefig('aer.png')
plt.show()
```

```python
color = raster.GetRGB()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(color)
plt.title('RGB Color')
plt.savefig('RGB.png')
plt.show()
```
